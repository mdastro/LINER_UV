# Functions

```python
def Kauffmann(x):
    return 0.61/(x-0.05)+1.3

def Kewley(x):
    return 0.61/(x-0.47)+1.19

def Stasinska(x):
    y = []
    for i in range(len(x)):
        y.append((-30.787+1.1358*x[i]+0.27297*x[i]**2)*math.tanh(5.7409*x[i])-31.093)
    return y

def Schawinski(x):
    return 1.05*x+0.45

def horizontal(x, a):
    return 0*x+a

def vertical(y, a):
    return 0*y+a
```
